```python
import nltk
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
nltk.download('punkt')
```

```python
def load_text(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return f.read()
avatar_text = load_text('avatarSpeeches.txt')
shake_text = load_text('shakeSonnets.txt')
print(avatar_text[:500])
print(shake_text[:500])
```

```python
avatar_tokens = nltk.word_tokenize(avatar_text.lower())
shake_tokens = nltk.word_tokenize(shake_text.lower())
print(avatar_tokens[:20])
print(shake_tokens[:20])
```

```python
def lexical_diversity(tokens):
    return len(set(tokens)) / len(tokens)

print("Lexical diversity of Avatar speeches:", lexical_diversity(avatar_tokens))
print("Lexical diversity of Shake sonnets:", lexical_diversity(shake_tokens))
```

```python
fd_avatar = FreqDist(avatar_tokens)
fd_shake = FreqDist(shake_tokens)
fd_avatar.plot(20, title="Avatar Speeches - Most Common Words")
fd_shake.plot(20, title="Shake Sonnets - Most Common Words")
```

```python
long_avatar_words = [w for w in avatar_tokens if len(w) > 10]
long_shake_words = [w for w in shake_tokens if len(w) > 10]
print("Long words in Avatar speeches:", set(long_avatar_words))
print("Long words in Shake sonnets:", set(long_shake_words))
```
